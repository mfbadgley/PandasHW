```python
# % and count of male players

#put only the male genders in a dataframe
male = pymoli_df.loc[pymoli_df["Gender"] == 'Male']
#group the dataframe by SN, so to eliminate multiple SN's
groupmale_df = male.groupby(['SN'])
#print number of rows in grouped dataframe of only Female Gender, unique 'SN' 
print('Male: '+ str(len(groupmale_df)))
percent_male= len(groupmale_df)/len(pymoli_df.groupby('SN'))
print("Percent Male: "+"{:.2%}".format(percent_male))

```

```python
# % and count of female players

#put only the female genders in a dataframe
female = pymoli_df.loc[pymoli_df["Gender"] == 'Female']
#group the dataframe by SN, so to eliminate multiple SN's
group_df = female.groupby(['SN'])
#print number of rows in grouped dataframe of only Female Gender, unique 'SN' 
print('Female: '+ str(len(group_df)))
percent_female= len(group_df)/len(pymoli_df.groupby('SN'))
print("Percent Female: "+"{:.2%}".format(percent_female))

```

```python
# % and count of other/Non-Disclosed 

#put only the other/non-disclosed genders in a dataframe
other = pymoli_df.loc[(pymoli_df["Gender"] != 'Female') & (pymoli_df["Gender"] != 'Male')]
#group the dataframe by SN, so to eliminate multiple SN's
groupother_df = other.groupby(['SN'])
#print number of rows in grouped dataframe of only Female Gender, unique 'SN' 
print('Other: '+ str(len(groupother_df)))
percent_other= len(groupother_df)/len(pymoli_df.groupby('SN'))
print("Percent Other: "+"{:.2%}".format(percent_other))
```

```python
#purchase count, by Gender
group_female_purchase = female.groupby(['Item ID'])
print('Female: '+ str(len(group_female_purchase)))
group_male_purchase = male.groupby(['Item ID'])
print('Male: '+ str(len(group_male_purchase)))
group_other_purchase = other.groupby(['Item ID'])
print('Other: '+ str(len(group_other_purchase)))
```

```python
#grouping by AGE 
#create the bins
bins =[6,10,14,18,22,26,30,34,38,42,46]

#Create the names for the bins
group_names = ['6 to 10', '10 to 14', '14 to 8', '18 to 22', '22 to 26', '26 to 30', '30 to 34', '34 to 38', '38 to 42', '42 to 46']

#Cut AGE and place the scores into bins
pd.cut(pymoli_df["Age"], bins, labels=group_names)
```

```python
pymoli_df["Age Summary"] = pd.cut(pymoli_df["Age"], bins, labels=group_names)
pymoli_df.head(20)
```

```python
#purchase count by age group
# Creating a group based off of the bins, count rows in each age bin
age_groups = pymoli_df.groupby("Age Summary")
age_groups['Item Name'].count()
```

```python
#Avg Purchase Price by Age Group
age_groups = pymoli_df.groupby("Age Summary")
age_groups['Price'].mean().round(2)
```

```python
#total purchase value by age group
age_groups = pymoli_df.groupby("Age Summary")
age_groups['Price'].sum().round(2)
```

```python
#Normalized values
age_groups = pymoli_df.groupby("Age Summary")
users = age_groups['SN'].nunique()
(age_groups['Price'].sum()/users).round(2)
```

```python
#identify top 5 spenders in the game by total purchase value; 
#list in a table by SN, Purchase Count, Average Purchase Price, Total Purchase Value
#Group by SN, find mean of Prices, put into dataframe
avg_purchase_price = pd.DataFrame(pymoli_df.groupby("SN")['Price'].mean())
#Group by SN, find sum of prices, put into dataframe
group_total_purchase = pd.DataFrame(pymoli_df.groupby("SN")['Price'].sum())
#group by SN, find count of prices, put into dataframe
group_purchase_count = pd.DataFrame(pymoli_df.groupby("SN")['Price'].count())
#inner join first two dataframes, returning matching values of both
new_df = pd.merge(avg_purchase_price, group_total_purchase, left_index = True, right_index = True)
#inner join the third to the first merged dataframe, returning matching values of both
final_new_df = pd.merge(new_df, group_purchase_count, left_index = True, right_index = True)
#rename the columns
final_new_df = final_new_df.rename(columns = {"Price_x":"Average Purchase Price","Price_y":"Total Purchase Value","Price":"Purchase Count"})  
#sort the dataframe, in place, on Total Purhase value
final_new_df.sort_values('Total Purchase Value', ascending = False, inplace = True)
#show the top 5 
final_new_df.head(5)                         
```

```python
#Identify the 5 most popular items by purchase count, then list (in a table):
 # * Item ID
#  * Item Name
#  * Purchase Count
#  * Item Price
#  * Total Purchase Value
    
#put item id into dataframe
item_name = pd.DataFrame(pymoli_df["Item Name"])
#put item price into dataframe
item_price = pd.DataFrame(pymoli_df["Price"])
#put item ID into dataframe
item_id = pd.DataFrame(pymoli_df['Item ID'])
#item_name = pd.DataFrame(pymoli_df['Item Name'])
#Group by item id, find sum of prices, put into dataframe
item_total_purchase = pd.DataFrame(pymoli_df.groupby("Item ID")['Price'].sum())
#group by item id, find count of prices, put into dataframe
item_purchase_count = pd.DataFrame(pymoli_df.groupby("Item ID")['Price'].count())
#inner join first two dataframes, returning matching values of both
new_df = pd.merge(item_name, item_total_purchase, left_index = True, right_index = True)
#inner join the third to the first merged dataframe, returning matching values of both
second_new_df = pd.merge(new_df, item_purchase_count, left_index = True, right_index = True)
#inner join the fourth to the 2nd merged dataframe, returning matching values of both
next_new_df = pd.merge(second_new_df, item_price, left_index = True, right_index = True)
final_new_df = pd.merge(next_new_df, item_id, left_index = True, right_index = True)
#rename the columns
final_new_df = final_new_df.rename(columns = {"Price_x":"Total Purchase Value","Price_y":"Purchase Count"})  
#sort the dataframe, in place, on Total Purhase value
final_new_df.sort_values('Purchase Count', ascending = False, inplace = True)
#show the top 5 
final_new_df.head(5)                         

#how to show the table below without the index column? 
```
